```python
# ==============================
# 1. IMPORT LIBRARY
# ==============================
import pandas as pd

# ==============================
# 2. LOAD DATASET
# ==============================
df = pd.read_excel("kelayakan-pendidikan-indonesia.xlsx")

# ==============================
# 3. CEK KONDISI AWAL DATASET
# ==============================
print("=== INFO DATASET AWAL ===")
df.info()

print("\n=== CONTOH DATA AWAL ===")
df.head()

print("\n=== DAFTAR KOLOM AWAL ===")
df.columns

# ==============================
# 4. NORMALISASI NAMA KOLOM (FIX ENCODING)
# ==============================
df.columns = (
    df.columns
    .str.strip()
    .str.replace("â‰¥", ">=", regex=False)
)

print("\n=== KOLOM SETELAH NORMALISASI ===")
df.columns

# ==============================
# 5. SELEKSI 4 PARAMETER UTAMA
# ==============================
selected_columns = [
    "Provinsi",
    "Siswa",
    "Putus Sekolah",
    "Kepala Sekolah dan Guru(>= S1)",
    "Ruang kelas(baik)"
]

df_clean = df[selected_columns]

# ==============================
# 6. OUTPUT DATASET SETELAH CLEANING
# ==============================
print("\n=== DATASET SETELAH DATA CLEANING ===")
df_clean.head()

# ==============================
# 7. CEK MISSING VALUE
# ==============================
print("\n=== CEK MISSING VALUE ===")
df_clean.isnull().sum()

# ==============================
# 8. CEK TIPE DATA
# ==============================
print("\n=== TIPE DATA ===")
df_clean.dtypes

# ==============================
# 9. UKURAN DATASET AKHIR
# ==============================
print("\n=== UKURAN DATASET AKHIR ===")
df_clean.shape

df_clean = df_clean.rename(columns={
    "Kepala Sekolah dan Guru(>= S1)": "Guru_Kepsek_S1_Keatas",
    "Ruang kelas(baik)": "Ruang_Kelas_Baik"
})

print("\n=== DATASET FINAL (NAMA KOLOM RAPI) ===")
df_clean.head()
```
